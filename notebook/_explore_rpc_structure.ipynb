```python
from processing.datareader import load_imu_data, load_radar_data, load_radar_config


datadir = "../dataset/town3_4"
rpc, frame_ids = load_radar_data(datadir)
sensor_transforms = load_radar_config(datadir)
imu_df = load_imu_data(f"{datadir}/imu_data.csv", ego_id=51)
```

```python
from processing.radarproc import RpcReplay

rpc_replay = RpcReplay(rpc, frame_ids, sensor_transforms, imu_df)
```

```python
from processing.clustering import RpcCluster

clustering = RpcCluster(rpc_replay, eps=2.0, min_samples=3, velocity_weight=10)
```
